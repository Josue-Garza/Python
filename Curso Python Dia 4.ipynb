##Funciones que llaman a otras funciones

```python
def saludar(nombre):
  return f"Hola {nombre}"

def despedirse(nombre):
  return f"Adiós, {nombre}"

def mensaje_completo(nombre):
  saludo = saludar(nombre)
  despedida = despedirse(nombre)
  return saludo + "\n" + despedida
print(mensaje_completo("Ana"))
```

```python
#Ahora supongamos que tenemos una heladería
def pedir_sabor():
  sabores = ["vainilla", "chocolate", "fresa"]
  print("Sabor disponible", ",", sabores)
  #.lower() hace minúsculas las palabras
  #.strip() elimina los espacios antes o después de las palabras "    vainilla" ---> "vainilla"
  sabor = input("Elige un sabor:").strip().lower()
  #Si sabor no está en sabores
  while sabor not in sabores:
    sabor = input("Elige un sabor (nuevamente):").strip().lower()
  return sabor

def pedir_cantidad():
  cantidad = int(input("¿Cuántas bolas de nieve quiere?: "))
  while cantidad <= 0:
    cantidad = int(input("¿Cuántas bolas de nieve quiere? (ingresa un valor válido): "))
  return cantidad

#Función madre (llama a otras funciones)
def pedir_helado():
  sabor = pedir_sabor()
  cant = pedir_cantidad()
  print(f"Pedido: {cant} bola(s) de {sabor}")
  return cant, sabor

total_pedidos = int(input("Cuántos clientes atenderás: "))
for _ in range(total_pedidos):
  pedir_helado()


```

##Funciones Lambda
Una función lambda es una función acortada. Para escribir una función de esta forma suele ser de esta manera:

nombreFuncion = lambda parámetros: acción

La diferencia entre las funciones lambda y las funciones normales es que las lambda pueden ser un poco más limitadas

```python
sumar = lambda x,y : x+y
print(sumar(2,3))
```

```python
descuentoLambda = lambda desc, precio: (precio * (1 - desc), (precio * desc))
nuevoPrecio = descuentoLambda(0.4, 500)
print(f"El precio a pagar es {nuevoPrecio[0]} y el descuento fue de {nuevoPrecio[1]}")
print(type(nuevoPrecio))
```

```python
#Calculadora: Suma, resta, multiplicación, división y potencia
def suma(a,b):
  return a+b

def resta(a,b):
  return a-b

def multiplicacion(a,b):
  return a*b

def division(a,b):
  if b == 0:
    return "Error: El divisor no puede ser cero"
  return a/b

def potencia(a,b):
  return a**b

def menu():
  print("----------MI CALCULADORA----------")
  print("1. Suma")
  print("2. Resta")
  print("3. Multiplicación")
  print("4. División")
  print("5. Potencia")
  print("6. Salir")

while True:
  menu()
  try:
    op = int(input("Seleccione una opción: "))
  except ValueError:
    print("Ingrese un valor numérico")
    continue

  if op < 1 or op > 6:
      print("Opción inválida")
      continue

  if op == 6:
    print("SALIENDO...")
    break

  try:
    a = float(input("Ingrese el primer número:"))
    b = float(input("Ingrese el segundo número:"))
  except ValueError:
    print("Ingrese un valor numérico")
    continue

  if op == 1:
    print("Resultado: ",suma(a,b))
  elif op == 2:
    print("Resultado: ",resta(a,b))
  elif op == 3:
    print("Resultado: ",multiplicacion(a,b))
  elif op == 4:
    print("Resultado: ",division(a,b))
  elif op == 5:
    print("Resultado: ",potencia(a,b))
```

```python
#EXTRA (opcional)
def promediosYmayor(promedios):
  promedio = 0
  for x in promedios:
    promedio += x
  promedio = promedio/len(promedios)
  return promedio, max(promedios)

promedios = []
continuar = ""

while True:
  if len(promedios) > 0:
    continuar = input("¿Desea ingresar otra calificación? ").lower()
    while continuar not in ("sí", "no"):
      continuar = input("Ingrese una respuesta válida: sí/no").lower()
    if continuar == "no":
      break
      promedios.clear()

  while True:
    try:
      calif = int(input("Indique la calificación: "))
      if calif < 0 or calif > 100:
        print("Ingrese un valor válido")
      else:
        promedios.append(calif)
        break
    except ValueError:
      print("Ingrese un valor numérico")
prom, mayor = promediosYmayor(promedios)
print("\nResultados:")
print("\nPromedio:", prom)
print("\nMayor:", mayor)
```
