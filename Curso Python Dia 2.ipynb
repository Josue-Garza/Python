##Elif
###Ahora hagamos un programa en el que preguntamos la calificación del 0 al 100, pero además queremos saber su equivalencia con el sistema de calificaciones de EEUU, basémonos en esto:
A+ = 100
A = 90-99
B = 80-89
C = 70-79
D = 60-69
F < 60



```python
cal = 100
if cal == 100:
  print("A+")
elif cal >= 90:
  print("A")
elif cal >= 80:
  print("B")
elif cal >=70:
  print("C")
elif cal >=60:
  print("D")
else:    #ya que ya no hay condición, el else cubre el resto de valores (cal<60)
  print("F")
```

##Operadores lógicos con condicionales

```python
#and,or,not
#Quieres un sándwich con todo
pan = True
queso = True
jamón = False
mayonesa = True
print(pan and queso and mayonesa and jamón)  #Hasta aquí va a correr un false porque no tenemos jamón
if not(pan and queso and mayonesa and jamón):
  print("No te puedes hacer el sándwich")
```

```python
cal = 100
if cal >= 0 and cal <= 100:
  if cal == 100:
    print("A+")
  elif cal >= 90:
    print("A")
  elif cal >= 80:
    print("B")
  elif cal >=70:
    print("C")
  elif cal >=60:
    print("D")
  else:    #ya que ya no hay condición, el else cubre el resto de valores (cal<60)
    print("F")
else:
  print("Error: la calificación no es válida. Debe de estar entre 0 y 100")
```

```python
#El módulo (%)
#El módulo es un operador matemático que devuelve el residuo de una división
#10 % 3 da 1, porque 10/3 es igual a 3 y sobra 1

num = 9
if num % 3 == 0 or num % 5 == 0:
  print(f"El número {num} es múltiplo de 3 O de 5.")
```

```python
#saber si el número es par o impar usando not

num = 5
if not num % 2 == 0:
  print(f"El número {num} es impar")
else:
  print(f"El número {num} es par")
```

##Ciclo FOR
###Los llamados ciclos for nos ayudan a realizar acciones durante una determinada cantidad de veces

```python
#Imprimir los números del 0 al 4
for i in range(5):   #Toma 5 como el número final, pero inicia SIEMPRE en 0
  print(i)
```

```python
for num in range (2,11):
  print(num)
```

```python
#También existe el range(inicio, fin, paso)
n = 10
for i in range(2,n+5,3):
  print(i)
```

```python
cadena = "Python Hola"
for letra_por_letra in cadena:
  print(letra_por_letra)
```

##While

```python
suma = 0
números = int(input("Ingresa el número que quieras y para salir escoge el 0: "))

while números != 0:
  suma += números
  números = int(input("Ingresa el número que quieras y para salir escoge el 0: "))

print(f"Tu resultado es {suma}")
```

```python
#Usar el While como For
numero = 0
while numero <=12:
  print(numero)
  numero += 1
```

```python
print("Juego: Adivina el Número")
numero_secreto = int(input("Jugador 1, introduce el número secreto (1-20): "))

intentos = 0
max_intentos = 4
while intentos < max_intentos:
  intento = int(input("Jugador 2, introduce un número del 1 al 20: "))
  intentos +=1

  if intento == numero_secreto:
    print("¡Correcto!")
    break
  else:
    print("¡Incorrecto! Vuelve a intentarlo")

#Opcional
if intentos == max_intentos and intento != numero_secreto:
  print("¡Se acabaron los intentos!")
  print(f"El múmero secreto era: {numero_secreto}")
```
