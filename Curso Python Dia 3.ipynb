#Estructuras de datos
Las estructuras de datos son formas de organizar información para trabajar con ella de manera eficiente.

- Listas => ordenadas, modificadas, permiten duplicados
- Tuplas => ordenadas, inmutables, permiten duplicados
- Sets => no ordenadas, sin duplicados
- Diccionarios => pares clave:valor

Nota:
- Pilas  (FILO => First in, Last Out)
- Colas   (FIFO => First in, First Out)
- Listas enlazadas (o doblemente enlazadas)

##Listas
Las listas son colecciones ordenadas que se pueden modificar y almacenan elementos de distintos tipos de datos.

```python
lista_de_compras = ["manzanas", "pan", "leche", "huevos", "mantequilla", "queso", "jamón"]
print(lista_de_compras)
```

```python
print(lista_de_compras[6])
```

```python
#Agregar un elemento a la lista .append()
lista_de_compras.append("sopa")
print(lista_de_compras)
```

```python
#Agregar un elemento en una posición específica con .insert()
lista_de_compras.insert(1,"plátano")
print(lista_de_compras)

#Si se vuelve a correr el programa se va a duplicar el elemento
```

```python
#Para eliminar un elemento de una lista existen el .remove() y el .pop()
lista_de_compras.remove("plátano")
print(lista_de_compras)
```

```python
lista_de_compras.pop() #Sin nada dentro del paréntesis borra el último elemento de la lista
lista_de_compras.pop(7) #Con un int dentro borra el elemento de la posición indicada INICIANDO DESDE CERO
#el .pop() retorna el valor eliminado que se puede guardar dentro de una variable
```

```python
#len() saca el tamaño de la lista
print(len(lista_de_compras))
for i in range(len(lista_de_compras)):
  print(f"{i+1} - {lista_de_compras[i]}")
```

```python
promedio = [55.6, 97.5, 23.6, 97.5, 72.4, 100]
print(promedio)
promedio.sort()    #.sort() Ordena una lista/arreglo de forma ascendente (valor y posición)
print(promedio)
#ordenar una lista de manera descendente
promedio.sort(reverse=True)
print(promedio)
```

```python
#Declaración y uso de matrices
calificaciones = [[100, 88, 93], [99, 56, 72]]
print(calificaciones)
i = 1
for cali_i in calificaciones:
  print(f"Alumno{i}")
  j = 1
  for cali_j in cali_i:
    print(f"Cali {j} : {cali_j}")
    j+=1
  j+=0
  i+=1
```

##Sets
Los sets son colecciones SIN orden y que NO permiten elementos duplicados

```python
nombres = {"Ángel", "Eliud", "Diana", "Josué", "Carmina", "Juan", "Juan"}
print(nombres)
print(type(nombres))
```

```python
#Unión
matutino = {20322, 22183, 13245, 86754}
vespertino = {20322, 14352, 67534, 22183}
#unión usando .union()
total_matriculas = matutino.union(vespertino)
print (total_matriculas)
#unión usando |
total_matriculas = matutino | vespertino
print(total_matriculas)
```

```python
#Intersección
ca = {1221, 458, 326, 524, 896, 21}
al = {998, 273847, 1221, 326, 587, 45}
#intersección usando .intersection()
ca_al = ca.intersection(al)
print(ca_al)
#Intersección usando &
ca_al = ca & al
print(ca_al)
```

```python
#Diferencia
ca = {1221, 458, 326, 524, 896, 21}
al = {998, 273847, 1221, 326, 587, 45}
#diferencia usando .difference()
solo_calculo = ca.difference(al)
print(solo_calculo)
#diferencia usando -
solo_calculo = ca - al
```

##Diccionarios
Son colecciones NO ordenadas de pares clave:valor

```python
alumno = {
    #clave    #valor
    "nombre" : "Josué Garza",
    "edad" : 18,
    "carrera": "IMTC",
    "semestre": 1,
    "promedio": 95,
    "calificaciones": [100, 89, 95, 99, 86]
}
print(alumno)
print(type(alumno))
```

```python
print("El alumno",alumno["nombre"], "está estudiando",alumno['carrera'])
```

```python
#Agregar nuevas áreas al diccionario
alumno["área preferida"] = "Programación aplicada a la robótica"
print(alumno)
```

```python
for valor in alumno:
  print(alumno[valor])
```

```python
for aspecto, valor in alumno.items():   #.items() devuelve una vista de todos los pares clave:valor del diccionario
  print(aspecto, ":", valor)
```

```python
#El método .clear también se puede usar aquí, no va a eliminar la variable (alumno), pero sí su información.
#Tenemos el .get() que nos ayuda a obtener el valor (value) de una cierta clave (key)
print(alumno.get("edad"))
```

```python
alumno = {
    #clave    #valor
    "nombre" : "Josué Garza",
    "edad" : 18,
    "carrera": "IMTC",
    "semestre": 1,
    "promedio": 95,
    "calificaciones": [100, 89, 95, 99, 86]
}
alumno_1 = alumno.items()
print(alumno_1)
print(type(alumno_1))
alumno_lista = list(alumno_1)
print(alumno_lista)
print(type(alumno_lista))
```

```python
#El método .values() sirve para obtener los valores de un diccionario, también se puede usar en la iteración
alumno_v = alumno.values()
print(type(alumno_v))
print(alumno_v)
```

```python
if "carrera" in alumno:
  print("Existe carrera")
else:
  print("No existe carrera")

if "dinero" in alumno:
  print("Alumno tiene dinero")
else:
  print("Alumno no tiene dinero")
```

```python
#.copy() puedes copiar todo el diccionario en una nueva variable
#.update() permite cambiar el diccionario, sea tan sólo un value, o bien, también un value y su key
alumno.update({"semestre":8})
print(alumno)
```

```python
#Código opcional
import random      #permite acceder a las funciones de random (en este caso random choice de las opciones que vamos a definir)
opciones = ["piedra", "papel", "tijeras"]

while True:
  while True:
    eleccion = input("\n Elige piedra, papel o tijeras (o escribe salir si ya no quieres jugar): ")
    if eleccion not in opciones and not eleccion == "salir":
      print("Opción no válida, intenta de nuevo")
    else:
      break
  if eleccion == "salir":
    print("¡Gracias por jugar!")
    break
  else:
     aleatoria = random.choice(opciones)
     print("La computadora eligió:",  aleatoria)

#Para determinar al ganador
  if eleccion == aleatoria:
    print("Empate")
  elif eleccion == "piedra" and aleatoria == "tijeras" or eleccion == "papel" and aleatoria == "piedra" or eleccion == "tijeras" and aleatoria == "papel":
    print("Ganaste")
  else:
    print("Perdiste")
```

```python
#lista de productos
productos = ["manzana", "plátano", "leche", "pan"]

#tupla de precios
precios = (10, 5, 20, 15)

#Diccionario de inventario: producto -> {"precio": x, "stock": y}
inventario = {productos[i]: {"precio": precios[i], "stock": 10} for i in range(len(productos))}
#<-- DICCIONARIO
#Clave: Nombre del producto
#Valor: Otro diccionario que incluye "precio" y "stock"

#Set de productos en oferta
ofertas = {"manzana", "pan"}   #<--- SET

#Mostrar productos
for p in productos:
  msg = "¡OFERTA!" if p in ofertas else "Grandes precios"
  info = inventario[p]
  print(f"{p.title()} - ${info["precio"]} - Stock: {info["stock"]} -> {msg}")

#Compra simple

compra = input("\n¿Qué producto quieres comprar? ")
if compra in inventario and inventario[compra]["stock"] > 0:
  inventario[compra]["stock"] -= 1
  print(f"Compraste 1 {compra}. Stock restante: {inventario[compra]["stock"]}")
else:
  print("Producto no disponible")
```
